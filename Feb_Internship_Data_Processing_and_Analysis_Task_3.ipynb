Problem Statement 1: Employee Performance Bonus Eligibility


```python
# Employee Performance Bonus Eligibility
employees = {"Ravi": 92,"Anita": 88,"Kiran": 92,"Suresh": 85}

# Step 1: Find the highest score
highest_score = max(employees.values())

# Step 2: Find all employees who have this highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Step 3: Display the result
print("Top Performers Eligible for Bonus:",
      ", ".join(top_performers),
      f"(Score: {highest_score})")
```

Problem Statement 2: Search Query Keyword Analysis               


```python
# Search Query Keyword Analysis
query = "Buy mobile phone buy phone online"
# Convert to lowercase
query = query.lower()

# Split sentence into words
words = query.split()

# Create empty dictionary to store word count
word_count = {}

# Count each word
for word in words:
    if word in word_count:
        word_count[word] = word_count[word] + 1
    else:
        word_count[word] = 1

# Print words that appear more than once
result = {}

for word in word_count:
    if word_count[word] > 1:
        result[word] = word_count[word]

print(result)
```

Problem Statement 3: Sensor Data Validation


```python
# Given sensor readings
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Create empty list to store valid readings
valid_readings = []

# Loop through the list using index
for i in range(len(sensor_readings)):

    # Check if the reading is even
    if sensor_readings[i] % 2 == 0:

        # Store as (hour_index, reading_value)
        valid_readings.append((i, sensor_readings[i]))

# Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

Problem Statement 4: Email Domain Usage Analysis


```python
# Email Domain Usage Analysis
# Given list of emails
emails = ["ravi@gmail.com","anita@yahoo.com","kiran@gmail.com","suresh@gmail.com","meena@yahoo.com"]

# Dictionary to store domain count
domain_count = {}

# Step 1: Count domains
for email in emails:
    # Split email to get domain
    parts = email.split("@")
    domain = parts[1]

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1
# Step 2: Calculate percentage
total_emails = len(emails)
for domain in domain_count:
    percentage = (domain_count[domain]/total_emails)*100
    print(domain+":",str(int(percentage))+"%")
```

Problem Statement 5: Sales Spike Detection


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
total = 0
for value in sales:
    total += value

average = total/len(sales)

# Approximate 30% (rounded for simplicity)
limit = average + 500   # 30% of 1700 ≈ 510, using 500 for easy logic

# Check spike days
for i in range(len(sales)):
    if sales[i] >= limit:
        print(f"Day {i+1}: {sales[i]}")
```

Problem Statement 6: Duplicate User ID Detection


```python
# Given list of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Step 1: Count each user ID
count_dict = {}

for user in user_ids:
    if user in count_dict:
        count_dict[user] += 1
    else:
        count_dict[user] = 1

# Step 2: Display only duplicates (count > 1)
for user in count_dict:
    if count_dict[user] > 1:
        print(user, "→", count_dict[user], "times")
```
