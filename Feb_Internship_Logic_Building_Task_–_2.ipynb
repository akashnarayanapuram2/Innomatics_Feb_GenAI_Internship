Problem 1: Unique Words in a Sentence              
Problem Statement                               
Given a sentence, identify all unique words and display:          
1. The count of unique words                 
2. The unique words themselves


```python
# user to enter a sentence
sentence = input("Enter sentence: ")

# Split the sentence into words
words = sentence.split()

# Convert list to set to remove duplicates
unique_words = set(words)

# Count unique words
count = len(unique_words)

# Display results
print("Unique words count:", count)
print("Unique words:", unique_words)
```

Problem 2: Highest Salary from Employee Data                   
Problem Statement                        
Store employee names and their salaries, and determine which employee has the highest salary.



```python
# Store employee names and their salaries
employees = { "Ravi": 75000, "Anita": 68000, "Kiran": 72000 }

# Assume the first employee has the highest salary
highest_name = list(employees.keys())[0]
highest_salary = employees[highest_name]

# Compare each salary
for name in employees:
    if employees[name] > highest_salary:
        highest_salary = employees[name]
        highest_name = name

# Display result
print("Highest Salary:", highest_name, "-", highest_salary)
```

Problem 3: Find Maximum and Minimum Values              
Problem Statement                            
Find the maximum and minimum values from a list of numbers.


```python
# Create a list of numbers
numbers = [45, 22, 89, 10, 66]

# Assume first number is both max and min
maximum = numbers[0]
minimum = numbers[0]

# Check each number
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Print results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

Problem 4: Count Products Above a Price Threshold                 
Problem Statement                            
Count how many products have a price greater than 1000.


```python
# Create a list of product prices
prices = [450, 1200, 899, 1500, 300]

# Counter variable
count = 0

# Check each price
for price in prices:
    if price > 1000:
        count = count + 1

# Print result
print("Products above 1000:", count)
```

Problem 5: Calculate Attendance Percentage                         
Problem Statement                          
Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).


```python
# List of attendance records
attendance = ["P", "P", "A", "P", "P"]

# Count present days
present_count = 0

for status in attendance:
    if status == "P":
        present_count = present_count + 1

# Total number of days
total_days = len(attendance)

# Calculate percentage
percentage = (present_count / total_days) * 100

# Print result
print("Attendance Percentage:", percentage)
```

Problem 10: Calculate Average Marks                     
Problem Statement                                  
Calculate the average marks of a student from a list of marks.


```python
# List of marks
marks = [70, 75, 80, 65]

# Calculate total marks
total = 0

for mark in marks:
    total = total + mark

# Count number of subjects
count = len(marks)

# Calculate average
average = total / count
print("Average Marks:", average)
```
