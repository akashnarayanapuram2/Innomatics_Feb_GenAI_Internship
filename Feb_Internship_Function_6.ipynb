Problem Statement 1:Smart Parking Lot Management System              
Design a function to manage a smart parking lot.



```python
def parking_system(capacity, vehicle_logs):

    parked = 0
    peak = 0

    # Count vehicles based on logs
    for entry in vehicle_logs:

        if entry == "IN":
            parked = parked + 1
        else:
            parked = parked - 1

        # Check highest parking at any time
        if parked > peak:
            peak = parked

    # Print current parked vehicles
    print("Currently Parked Vehicles:", parked)

    # Check capacity
    if parked > capacity:
        print("Parking Status: Full / Exceeded")
    else:
        print("Parking Status: Available")

# Sample Data
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parking_system(capacity, logs)
```

Problem Statement 2:Online Food Delivery Time Estimator                    

Create a function that estimates delivery time based on:                 
Distance (km)                                               
Weather condition                               
Traffic level                                 
Apply delays dynamically and display final ETA.


```python
def estimate_delivery_time(distance, traffic, weather):

    # Base time
    time = 5

    # Add time based on distance
    time = time + (distance * 5)

    # Traffic condition
    if traffic == "High":
        time = time + 10
    elif traffic == "Medium":
        time = time + 5

    # Weather condition
    if weather == "Rainy":
        time = time + 10
    elif weather == "Cloudy":
        time = time + 5

    print("Estimated Delivery Time:", time, "minutes")

# Sample Input
estimate_delivery_time(6, "High", "Rainy")
```

 Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

                 
Build a function that analyzes seat booking data and:                     
Calculates occupancy percentage                                
Determines if show is Housefull                   
Suggests opening additional shows                            

```python
def analyze_seat_occupancy(total_seats, booked_seats):

    # Count booked seats using loop
    booked_count = 0

    for seat in booked_seats:
        booked_count = booked_count + 1

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
        print("Suggestion: Open Additional Show")

    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# Sample Input
total = 200

# Creating list with 150 booked seats
booked = []
for i in range(150):
    booked.append(1)

# Call function
analyze_seat_occupancy(total, booked)
```

 Problem Statement 4:Cloud Server Load Classification System
         
Create a function to classify server load based on CPU usage readings.    
Rules:                                          
Average CPU < 50% → Normal                           
50%–80% → Warning                      
80% → Critical                

```python
def classify_server_load(cpu_readings):

    total = 0

    # Calculate total using loop
    for value in cpu_readings:
        total = total + value

    # Calculate average
    average = total / len(cpu_readings)

    print("Average CPU Load:", int(average), "%")

    # Classify server status
    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu_data = [45, 60, 70, 85, 90]

# Call function
classify_server_load(cpu_data)
```

 Problem Statement 5:Smart Classroom Resource Usage Monitor

Design a function that tracks usage of classroom resources (projector, AC, lights) and identifies overuse patterns.

```python
def check_resource_usage(resource_data):

    overused_list = []

    # Loop through dictionary
    for resource in resource_data:

        hours = resource_data[resource]

        # Check if usage is more than 8 hours
        if hours > 8:
            overused_list.append(resource)

    # Print overused resources
    if len(overused_list) > 0:
        print("Overused Resources:", ", ".join(overused_list))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

# Call function
check_resource_usage(usage)
```

 Problem Statement 6: Online Event Registration Capacity Controller         

Create a function that manages event registrations by:                 
Tracking registrations                           
Preventing overbooking                            
Triggering waitlist mode                                 


```python
def manage_registrations(capacity, total_registrations):

    confirmed = 0

    # Count registrations up to capacity
    for i in range(total_registrations):

        if confirmed < capacity:
            confirmed = confirmed + 1

    # Calculate waitlisted users
    waitlisted = total_registrations - confirmed

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    # Registration status
    if total_registrations >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# Sample Input
event_capacity = 100
registrations = 105

# Call function
manage_registrations(event_capacity, registrations)
```
