Problem 1: Social Media – Post Engagement Analyzer                
Problem Statement                               
Create a function to analyze engagement on a social media post.                


```python
# Function to analyze post engagement
def analyze_post(likes_list):

    total_likes = 0

    # Calculate total likes using loop
    for likes in likes_list:
        total_likes += likes

    # Print total likes
    print("Total Likes:", total_likes)

    # Check if post is viral
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example likes data (likes from different hours or social media)
likes_data = [200, 300, 150, 250, 150]

# Call function
analyze_post(likes_data)
```

Problem 3: Agriculture – Rainfall Adequacy Checker                     
Problem Statement                                      
Create a function that analyzes seasonal rainfall data.                     


```python
# Function to check rainfall adequacy
def check_rainfall(rainfall_data, required_level):

    total = 0

    # Calculate total rainfall using loop
    for value in rainfall_data:
        total += value

    # Calculate average rainfall
    average = total / len(rainfall_data)

    print("Average Rainfall:", int(average))

    # Check if rainfall is adequate
    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")

# Example seasonal rainfall data
rainfall_list = [70, 75, 68, 80, 67]
required_rainfall = 70

# Call function
check_rainfall(rainfall_list, required_rainfall)
```

Problem 4: Social Media – Duplicate Account Detection                    
Problem Statement                                             
Create a function to detect duplicate usernames in a system.                   


```python
# Function to check duplicate usernames
def check_duplicates(usernames):

    unique_users = set(usernames)

    # If length of set is smaller, duplicates exist
    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example username list
user_list = ["akash123", "ravi01", "meena07", "akash123"]

# Call function
check_duplicates(user_list)
```

Problem 9: E-Learning – Student Attendance Eligibility System             
Problem Statement                                    
Create a function to calculate attendance percentage and check exam eligibility.


```python
# Function to calculate attendance and check eligibility
def check_exam_eligibility(attendance_list):

    present_count = 0

    # Count present days using loop
    for day in attendance_list:
        if day == "P":
            present_count += 1

    # Calculate attendance percentage
    percentage = (present_count / len(attendance_list)) * 100

    print("Attendance Percentage:", percentage)

    # Check eligibility
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example attendance data
attendance = ["P", "P", "A", "P", "P"]

# Call function
check_exam_eligibility(attendance)
```

Problem 10: Smart Electricity Bill Analyzer                    
Problem Statement                                      
You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units consumed.

```python
# Function to calculate electricity bill
def calculate_bill(units):

    bill = 0

    # Slab 1: First 100 units
    if units <= 100:
        bill = units * 3

    # Slab 2: 101–200 units
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)

    # Slab 3: Above 200 units
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage Classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status

# Example usage
units_consumed = 250

total_bill, usage_status = calculate_bill(units_consumed)

print("Total Bill:", total_bill)
print("Usage Status:", usage_status)
```
