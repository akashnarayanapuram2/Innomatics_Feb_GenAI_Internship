1. User Login Check                            
Problem Statement:                              
Given a username and password, check whether login is successful.       
username = "admin"                     
password = "1234"             



```python
# Correct username and password
correct_username = "admin"
correct_password = "1234"

# Take input from user
username = input("Enter username: ")
password = input("Enter password: ")

# Check login details
if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

2. Pass / Fail Analyzer                            
Problem Statement:                         
Analyze student results.                                   
marks = [45, 78, 90, 33, 60]



```python
# List of student marks
marks = [45, 78, 90, 33, 60]

# Counters
pass_count = 0
fail_count = 0

# Check each mark
for mark in marks:
    if mark >= 50:
        pass_count = pass_count + 1
    else:
        fail_count = fail_count + 1

# Print results
print("Total Students:", len(marks))
print("Pass Students:", pass_count)
print("Fail Students:", fail_count)
```

3. Simple Data Cleaner                              
Problem Statement:                                     
Clean and standardize user names.                                     
names = [" Alice ", "bob", " CHARLIE "]               



```python
# List of names
names = [" Alice ", "bob", " CHARLIE "]

# Empty list to store cleaned names
cleaned_names = []

# Clean each name
for name in names:
    # Remove spaces and convert to lowercase
    cleaned = name.strip().lower()
    cleaned_names.append(cleaned)

# Print cleaned list
print("Cleaned Names:", cleaned_names)
```

4. Message Length Analyzer                         
Problem Statement:                            
Analyze message sizes.                                      
messages = ["Hi", "Welcome to the platform", "OK"]


```python
# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Check each message
for message in messages:
    length = len(message)
    print("Message:", message)
    print("Length:", length)

    if length > 10:
        print("This message is longer than 10 characters")

    print()  # Empty line for clear output
```

5. Error Message Detector                  
Problem Statement:                            
Detect error messages from system logs.                                   
logs = ["INFO", "ERROR", "WARNING", "ERROR"]



```python
# List of system logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Variable to count errors
error_count = 0

# Loop through each log entry
for log in logs:
    # Check if the log is "ERROR"
    if log == "ERROR":
        error_count = error_count + 1

# Print total number of errors
print("Total Error Count:", error_count)
```
